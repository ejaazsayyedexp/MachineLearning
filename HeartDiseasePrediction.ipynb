```python
import pandas as pd
import numpy as np
```

```python
heartData = pd.read_csv('./datasets_479954_897254_heart.csv')
```

```python
heartData.describe()
#Column Wise description
#1. Age
#2. Sex
#3. cp - Chest Pain
#4. trestbps - Resting Blood Pressure
#5. chol - Cholestrol
#6. fbs - fasting blood sugar
#7. restexg - resting ecg results
#8. thalach - Maximum Heart rate achieved
#9. exang - Exercise induced angina
#10. oldpeak - ST depression induced by exercise relative to rest
#11. Slope: Slope of the peak exercise ST segment
#12. Ca: Number of major vessels colored by flouroscopy (0 - 3)
#13. Thal: (3 = normal; 6 = fixed defect; 7 = reversable defect)
#14. target: AHD - Diagnosis of heart disease (1 = yes; 0 = no)
```

```python
heartData.tail(20)
```

```python
features = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']
features
```

```python
X = heartData[features]
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
y = heartData.target
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 0)
```

```python
heartDataModel = DecisionTreeRegressor(random_state = 1)
heartDataModel.fit(X_train,y_train)
```

```python
predictions = heartDataModel.predict(X_test)
err = mean_absolute_error(predictions,y_test)
err
```
